Array Operations
 Create two 2D arrays of shape (3, 3) and perform element-wise addition and
multiplication.
 Create a 4x4 array of random integers between 10 and 50. Find the maximum and
minimum values in the array.

```python
arr=np.arange(9).reshape(3,3)
print(arr)
arr2=np.arange(9).reshape(3,3)
print(arr2)
arr3=arr+arr2
print(arr3)
arr4=arr*arr2
print(arr4)
arr5=np.random.randint(10,50,size=(4,4))
print(arr5)
print(arr5.max())
print(arr5.min())
```

Statistical Analysis

Advanced Indexing
 Create a 4x4 matrix and use boolean indexing to set all elements greater than 5 to -1.
 Generate a 2D array of shape (5, 5) and use advanced indexing to extract specific rows
and columns based on conditions.

```python
import numpy as np

# Step 1
arr = np.arange(16).reshape(4, 4)
print("Original arr:")
print(arr)

# Step 2
arr[arr > 5] = -1
print("Modified arr:")
print(arr)

# Step 3
arr2 = np.arange(25).reshape(5, 5)
print("arr2:")
print(arr2)

# Step 4
# Correcting the if statement
if (arr2[1:] > 5).any():
    print("Sub-array where elements are greater than 5:")
    print(arr2[1:])
else:
    print("no")
```
